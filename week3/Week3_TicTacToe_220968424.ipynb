```python
from easyAI import TwoPlayerGame
from easyAI.Player import Human_Player

class TicTacToe( TwoPlayerGame ):
    
    """ 
        The board positions are numbered as follows:
            1 2 3
            4 5 6
            7 8 9
    """    

    def __init__(self, players):
        self.players = players
        self.board = [0 for i in range(9)]
        self.current_player = 1 # player 1 starts.
    
    def possible_moves(self):
        return [i+1 for i,e in enumerate(self.board) if e==0]
    
    def make_move(self, move):
        self.board[int(move)-1] = self.current_player

    def unmake_move(self, move): # optional method (speeds up the AI)
        self.board[int(move)-1] = 0
    
    def lose(self):
        opponent = 3 - self.current_player
        return any([all([self.board[c - 1] == opponent for c in line]) 
                    for line in [[1, 2, 3], [4, 5, 6], [7, 8, 9],  # horizontal
                                 [1, 4, 7], [2, 5, 8], [3, 6, 9],  # vertical
                                 [1, 5, 9], [3, 5, 7]]])  # diagonal
        
    def is_over(self):
        return (self.possible_moves() == []) or self.lose()
        
    def show(self):
        print()
        for row in range(3):
            row_values = [ [' ', 'O', 'X'][self.board[3*row + col]] for col in range(3) ]
            print(" | ".join(row_values))
            if row < 2:
                print("---------")
        print()

                 
    def scoring(self):
        return -100 if self.lose() else 0
```

```python
from easyAI import AI_Player, Negamax
from easyAI.AI import solve_with_depth_first_search

if __name__ == "__main__":
    
    ai_algo = Negamax(6)
    
    game = TicTacToe([Human_Player(), AI_Player(ai_algo)])
    game.play()
    
    result = solve_with_depth_first_search(game, win_score=100)
    
    if result == 1:
        print("DFS Result: Certain victory for the first player.")
    elif result == 0:
        print("DFS Result: Draw or insufficient depth for a conclusive result.")
    elif result == -1:
        print("DFS Result: Certain defeat for the first player.")
```

```python
from easyAI.AI import solve_with_iterative_deepening

if __name__ == "__main__":
    
    ai_algo = Negamax(6)
    
    game = TicTacToe([ Human_Player(), AI_Player(ai_algo)])
    game.play()
    
    ai_depths = range(1, 10) 
    win_score = 100
    result, depth, moves = solve_with_iterative_deepening(game, ai_depths, win_score)
    
    print()
    print(f"Iterative Deepening Result: {result}, Depth: {depth}, Moves: {moves}")
```
